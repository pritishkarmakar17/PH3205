```python
import numpy as np
import matplotlib.pyplot as plt
```

# Q1

```python
s = 0.2; x0 = 3

f = lambda x: (1/np.sqrt(2*np.pi*s**2))*np.exp((-(x-x0)**2)/(2*s**2))
```

```python
V0 = 40

def V(x):
    if (x<=7) and (x>=5):
        return V0
    else:
        return 0
```

```python
N = 300
dt = 0.01

xs, h = np.linspace(-10.0,10.0,N,retstep=True,endpoint= True)

psi0 = f(xs)

Vs = np.array ([V(i) for i in xs])

plt.plot(xs,Vs)
```

```python
plt.plot(xs,psi0)
```

```python
kt = 2j/dt

A = np.diag(-2 +(kt-Vs)*h**2) + np.eye(N,k=1) + np.eye(N,k=-1)
B =  np.linalg.inv(A)
U =  B*(4*1j*h**2)/dt - np.eye(N)
```

```python
t = 0
tf = 2

k = 0
psi = psi0

while t<tf:
    psi = np.dot(U,psi)
    t += dt
    if k%40==0:
        plt.plot(xs,np.real(psi),label=str(round(t,2)))
        plt.legend()
    k+=1
plt.plot(xs,Vs/np.max(Vs), ":")
plt.show()
```
